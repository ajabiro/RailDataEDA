```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_parquet('rail_data_cleaned_20230728132300.parquet')
```

```python
df.head()
```

```python
# Analzying and visualizing peak hours of train activity
```

```python
# Extract hour from arrival and departure times
df['arrival_hour'] = df['actual_arrival_time'].dt.hour
df['departure_hour'] = df['actual_departure_time'].dt.hour
```

```python
# Group by hour and count arrivals and departures
arrival_counts = df.groupby('arrival_hour').size()
departure_counts = df.groupby('departure_hour').size()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

plt.plot(arrival_counts.index, arrival_counts.values, label='Arrivals', marker='o')
plt.plot(departure_counts.index, departure_counts.values, label='Departures', marker='o')

plt.title('Train Activity by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trains')
plt.xticks(range(24))
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Analyzing Routes - finding the most popular routes
popular_routes = df['route_id'].value_counts()
popular_trains = df['unique_id'].value_counts()
popular_stations = df['train_platform'].value_counts()
```

```python
top_15_routes = popular_routes.head(15)
print(top_15_routes)
```

```python
top_15_stations = popular_stations.head(15)
print(top_15_stations)
```

```python
top_15_trains= popular_trains.head(15)
print(top_15_trains)
```

```python
# # sorting the routes in descending order most to least popular
# popular_routes = popular_routes.sort_values(ascending=False)
# popular_trains = popular_trains.sort_values(ascending=False)
# popular_stations = popular_stations.sort_values(ascending=False)
```

```python
top_15_routes.plot(kind='bar')
plt.xlabel('Route ID')
plt.ylabel('Frequency')
plt.title('Top 15 Most Popular Routes')
plt.show()
```

```python
top_15_trains.plot(kind='bar')
plt.xlabel('Train ID')
plt.ylabel('Frequency')
plt.title('Top 15 Most Popular Trains')
plt.show()
```

```python
top_15_stations.plot(kind='bar')
plt.xlabel('Station Name')
plt.ylabel('Frequency')
plt.title('Top 15 Most Popular Stations')
plt.show()
```

```python
# Determining the busiest station on a specific day (7/28/2023)
```

```python
specific_date = df[df['service_start_date'] == '2023-07-28']
```

```python
station_counts = specific_date['train_platform'].value_counts()
```

```python
sorted_stations = station_counts.sort_values(ascending=False)
```

```python
top_10_stations = sorted_stations.head(10)
```

```python
top_10_stations
```

```python
top_10_stations.plot(kind='bar')
plt.xlabel('Station Name')
plt.ylabel('Frequency')
plt.title('Top 10 Visited Stations in a Day (2023-07-28)')
plt.show()
```

```python
top_10_labels = top_10_stations.index.tolist()
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_10_stations, labels=top_10_labels, autopct='%1.1f%%')
plt.title('Top 10 Most Visited Stations on 7/28/2023')
plt.show()
```

```python
# Analyzing Train Length to see if there is a correlation between the length of train and its arrival/departure time.
# - Do longer trains tend to take more time at each station?
```

```python
df.train_length.unique()
```

```python
delayed_arrivals = df[df['is_delayed_arrival'] == True]['train_length'].value_counts().sort_values(ascending=True)
```

```python
print(delayed_arrivals)
```

```python
delayed_departures = df[df['is_delayed_departure'] == True]['train_length'].value_counts().sort_values(ascending=True)
```

```python
delayed_departures
```

```python
delayed_count_df = pd.DataFrame({'Delayed Arrivals': delayed_arrivals, 'Delayed Departures': delayed_departures})
```

```python
delayed_count_df.plot(kind='bar')
plt.xlabel('Train Length (Number of Cars)')
plt.ylabel('Delayed Count')
plt.title('Delayed Arrivals and Departures by Train Length')
plt.show()
```

```python
non_delayed_arrivals = df[df['is_delayed_arrival'] == False]['train_length'].value_counts().sort_values(ascending=True)
```

```python
non_delayed_departures = df[df['is_delayed_departure'] == False]['train_length'].value_counts().sort_values(ascending=True)
```

```python
non_delayed_count_df = pd.DataFrame({'Delayed Arrivals': non_delayed_arrivals, 'Delayed Departures': non_delayed_departures})
```

```python
non_delayed_count_df.plot(kind='bar')
plt.xlabel('Train Length (Number of Cars)')
plt.ylabel('Count')
plt.title('Non Delayed Arrivals and Departures by Train Length')
plt.show()
```

```python
# count_df=count_df.sort_index()
```

```python
count_df = pd.DataFrame({'Delayed Arrivals': delayed_arrivals, 'Delayed Departures': delayed_departures, 'Non Delayed Arrivals': non_delayed_arrivals, 'Non Delayed Departures': non_delayed_departures})
```

```python
count_df.plot(kind='bar')
plt.xlabel('Train Length (Number of Cars)')
plt.ylabel('Count')
plt.title('Arrivals and Departures by Train Length')
plt.show()
```
